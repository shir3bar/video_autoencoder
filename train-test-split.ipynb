```python
import cv2
import matplotlib.pyplot as plt
import os
import numpy as np
import pandas as pd
import random
from AuxiliaryFunctions import split_clip
from ast import literal_eval as make_tuple
```

```python
root='G:/My Drive/PhD/Fishbase/SwimVSFeed/New_DB/'
root_folders = os.listdir(root)
root_folders = [folder for folder in root_folders if not (folder.startswith('test') or folder.startswith('train')) ][:3]
train_ratio = 0.75
```

```python

for folder in root_folders:
    files = os.listdir(os.path.join(root,folder))
    new_files = [folder+'_'+f for f in files]
    num_files = len(files)
    num_train = round(num_files*train_ratio)
    shuffled_idxs = np.random.permutation(range(num_files))
    train_idxs = shuffled_idxs[:num_train]
    test_idxs = shuffled_idxs[num_train:]
    for i in range(num_files):
        if i in train_idxs:
            new_folder = 'train_500px'
        else:
            new_folder = 'validation_500px'
        os.rename(os.path.join(root,folder,files[i]),os.path.join(root,new_folder,new_files[i]))
    print(f'num files in folder: {num_files}, train idxs: {len(train_idxs)}, test idx: {len(test_idxs)}')
            
    
```
